```python
import pandas as pd
from docxtpl import DocxTemplate

data = pd.read_excel(r'C:\Users\qeto\Desktop\moxaliseebi.xlsx')

for index, row in data.iterrows():
    doc = DocxTemplate(r'C:\Users\qeto\Desktop\EMPLOYEE OF THE MONTH.docx')
    context = {
        "NAME": row['NAME'],
        "EVENT": row['EVENT'],
        "DATE": row['DATE'].strftime("%m.%d.%Y")
    }
    doc.render(context)
    filename = f"sertifikati_{row['NAME']}.docx"
    doc.save(filename)
```
